###### Take each csv file , split that csv file into multiple categories (example csv files are added in the repo) 


###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


###### if  10000<codekata score<15000   (Reached_expectations.csv)



###### if  7000<codekata score<10000   (Needs_Improvement.csv)


###### if  codekate score < 7000        (Unsatisfactory.csv)

```python
#College1
College1_Exceeded_expectations=r"C:\Users\sanka\Downloads\College1_Exceeded_expectations.csv"
College1_Reached_expectations=r"C:\Users\sanka\Downloads\College1_Reached_expectations.csv"
College1_Needs_Improvement=r"C:\Users\sanka\Downloads\College1_Needs_Improvement.csv"
College1_Unsatisfactory=r"C:\Users\sanka\Downloads\College1_Unsatisfactory.csv"
#College1
College2_Exceeded_expectations=r"C:\Users\sanka\Downloads\College1_Exceeded_expectations.csv"
College2_Reached_expectations=r"C:\Users\sanka\Downloads\College2_Reached_expectations.csv"
College2_Needs_Improvement=r"C:\Users\sanka\Downloads\College2_Needs_Improvement.csv"
College2_Unsatisfactory=r"C:\Users\sanka\Downloads\College2_Unsatisfactory.csv"
```

```python
#College1
ee1=c1.loc[c1["CodeKata Score"]>15000]
re1=c1.loc[c1["CodeKata Score"].between(10000,15000)]
ni1=c1.loc[c1["CodeKata Score"].between(7000,10000)]
us1=c1.loc[c1["CodeKata Score"]<7000]

ee1.to_csv(College1_Exceeded_expectations,index=False)
re1.to_csv(College1_Reached_expectations,index=False)
ni1.to_csv(College1_Needs_Improvement,index=False)
us1.to_csv(College1_Unsatisfactory,index=False)


#College2
ee2=c2.loc[c1["CodeKata Score"]>15000]
re2=c2.loc[c1["CodeKata Score"].between(10000,15000)]
ni2=c2.loc[c1["CodeKata Score"].between(7000,10000)]
us2=c2.loc[c1["CodeKata Score"]<7000]

ee2.to_csv(College2_Exceeded_expectations,index=False)
re2.to_csv(College2_Reached_expectations,index=False)
ni2.to_csv(College2_Needs_Improvement,index=False)
us2.to_csv(College2_Unsatisfactory,index=False)
```
